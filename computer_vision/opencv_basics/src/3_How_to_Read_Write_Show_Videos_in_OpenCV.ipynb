cv2.VideoCapture()  
* either takes file name ('filename.avi') you want to read as an argument 
* or you have to provide device index of your camera from which to want to read, by deafault this index is either "0" or "-1".
* In case you are having multiple cameras u can use index to denote which u want 
                    i.e. cap = cv2.VideoCapture(1)--> for 1st camaera
                         cap = cv2.VideoCapture(2)--> for 2nd camera

```python
cap = cv2.VideoCapture(0)

while(True):                        # capturing frames till while is true
    
    ret, frame = cap.read()         # ret will return True/False, return true if frame is available 

    cv2.imshow('frame_capture', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()              
cv2.destroyAllWindows()
```

### To read the grey scale video

```python
cap = cv2.VideoCapture(0)

while(True):                     
    
    ret, frame = cap.read()          
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)           # default order for color is BGR

    cv2.imshow('frame_capture', gray)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()              
cv2.destroyAllWindows()
```

### what if you provide wrong index or wrong file name to the argument of cv2.VideoCapture() function

```python
cap = cv2.VideoCapture(45)
print(cap.isOpened())                             # results in "False" since index = 45 is not a valid one.

while(cap.isOpened()):                     
    
    ret, frame = cap.read()              
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)           # default order for color is BGR
    cv2.imshow('frame_capture', gray)   
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()              
cv2.destroyAllWindows()
```

## How to save captured video

```python
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')                                     # or ('X','V','I','D')

out = cv2.VideoWriter('temp_file/output_video_capture.avi', fourcc, 20.0, (640,480))  # 20--> 20 frames/sec, (640,480)--> frame size

print(cap.isOpened())
while(cap.isOpened()):
    ret, frame = cap.read()                    
    if ret == True:                                                         # if ret==True then only we save our frame
       print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))                             # print the width of the frame being captured
       print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

       out.write(frame)                     # our video will save in BGR mode, bcz we are saving it before the conversion

       gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
       cv2.imshow('frame', gray)

       if cv2.waitKey(1) & 0xFF == ord('q'):
         break
    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
